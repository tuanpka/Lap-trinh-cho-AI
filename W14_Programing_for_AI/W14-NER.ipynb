Sử dụng bộ dữ liệu này: https://groups.csail.mit.edu/sls/downloads/movie/

Training: engtrain.bio

Testing: engtest.bio

Xây dựng và so sánh các mô hình học máy trong scikit-learn cho bài toán NER.


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

# Read the data
def read_data(file_path):
    with open(file_path, 'r') as file:
        data = file.read().split('\n')
    sentences, labels = [], []
    sentence, label = [], []
    for line in data:
        if line.strip() == '':
            if sentence:
                sentences.append(' '.join(sentence))
                labels.append(' '.join(label))
                sentence, label = [], []
        else:
            word, tag = line.split()
            sentence.append(word)
            label.append(tag)
    return sentences, labels

train_sentences, train_labels = read_data('/content/engtrain.bio.txt')
test_sentences, test_labels = read_data('/content/engtest.bio.txt')

# Convert labels to a single list for training
def flatten_labels(labels):
    return [tag for label in labels for tag in label.split()]

flat_train_labels = flatten_labels(train_labels)
flat_test_labels = flatten_labels(test_labels)
```

```python
# Convert sentences to a single list for vectorization
flat_train_sentences = [word for sentence in train_sentences for word in sentence.split()]
flat_test_sentences = [word for sentence in test_sentences for word in sentence.split()]

# Vectorize the data
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(flat_train_sentences)
X_test = vectorizer.transform(flat_test_sentences)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB

# Train Decision Tree model
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, flat_train_labels)

# Train Naive Bayes model
nb_model = MultinomialNB()
nb_model.fit(X_train, flat_train_labels)
```

```python
from sklearn.metrics import classification_report, accuracy_score

# Predict using Decision Tree model
dt_predictions = dt_model.predict(X_test)

# Predict using Naive Bayes model
nb_predictions = nb_model.predict(X_test)

# Evaluate the models
print("Decision Tree Classification Report:")
print(classification_report(flat_test_labels, dt_predictions, zero_division=0))

print("Naive Bayes Classification Report:")
print(classification_report(flat_test_labels, nb_predictions, zero_division=0))

# Calculate accuracy scores
dt_accuracy = accuracy_score(flat_test_labels, dt_predictions)
nb_accuracy = accuracy_score(flat_test_labels, nb_predictions)

print(f"Decision Tree Accuracy: {dt_accuracy}")
print(f"Naive Bayes Accuracy: {nb_accuracy}")
```
