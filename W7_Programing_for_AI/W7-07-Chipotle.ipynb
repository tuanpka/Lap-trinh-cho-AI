### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv'
    
chipo = pd.read_csv(url, sep = '\t')
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
x = chipo.item_name

letter_counts = Counter(x)

df = pd.DataFrame.from_dict(letter_counts, orient='index')

df = df[0].sort_values(ascending = True)[45:50]

df.plot(kind='bar')

plt.xlabel('Items')
plt.ylabel('Number of Times Ordered')
plt.title('Most ordered Chipotle\'s Items')

#plot
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
# List of prices
chipo.item_price = [float(value[1:-1]) for value in chipo.item_price] # strip the dollar sign and trailing space

orders = chipo.groupby('order_id').sum()

# Creates scatterplot
plt.scatter(x = orders.item_price, y = orders.quantity, s = 50, c = 'green')

# Title
plt.xlabel('Order Price')
plt.ylabel('Items ordered')
plt.title('Number of items ordered per order price')
plt.ylim(0)
```

### BONUS: Create a question and a graph to answer your own question.

### Question: Create top 5 most ordered items at this restaurant and their total sales

```python
# Calculate the total sales for each item
chipo['total_sales'] = chipo['quantity'] * chipo['item_price']

# Group by item name and sum up the total sales and total quantity for each item
item_sales = chipo.groupby('item_name').agg({'total_sales': 'sum', 'quantity': 'sum'}).reset_index()

# Sort the items by the total sales in descending order
top_items = item_sales.sort_values(by='total_sales', ascending=False).head(5)

# Plotting
plt.figure(figsize=(8, 4))
plt.barh(top_items['item_name'], top_items['total_sales'], color='skyblue')
plt.xlabel('Total Sales ($)')
plt.ylabel('Item Name')
plt.title('Top 5 Most Ordered Items and Their Sales')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest sales on top
plt.show()
```
